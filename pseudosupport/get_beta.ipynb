```python
import os,sys
import pandas as pd
```

```python
demographics_folder = "2021_GCP_CED_for_AUS_short-header/2021 Census GCP Commonwealth Electroral Division for AUS"
demographic_files = os.listdir(demographics_folder)
demographic_db = pd.read_csv(f"{demographics_folder}/{demographic_files[0]}",index_col="CED_CODE_2021")
cedcodemapper = pd.read_csv("CEDcodemapper.csv",index_col="CED_CODE_2021")
demographic_db.head()
```

```python
for file in demographic_files[1:]:
    demographic_db = pd.concat(
        [demographic_db,
        pd.read_csv(f"{demographics_folder}/{demographic_files[0]}",index_col="CED_CODE_2021")],axis=1
    )
demographic_db = pd.merge(cedcodemapper,demographic_db, left_index=True, right_index=True, how='inner').set_index("DivisionNm")

demographic_db = demographic_db.T.drop_duplicates().T
```

```python
primaryvotes = pd.read_csv("HouseFirstPrefsByCandidateByVoteTypeDownload-27966.csv",header=1)
primaryvotes
```

```python
teal_candidates = pd.read_csv("teal.csv")
teal_candidates.head()
# teal_candidates['Surname'] = teal_candidates['Candidate'].apply(lambda x: ' '.join(x.split()[1:]).upper())
# teal_candidates['GivenNm'] = teal_candidates['Candidate'].apply(lambda x: ' '.join(x.split()[0:1]))
```

```python
parties_collected = ['UAPP', 'ON', 'ALP', 'IND', 'GRN', 'LP','NP','CLP','LNP','XEN', 'KAP']
party_mappers = {
    'UAPP':'UAP', 
    'ON':'ONP',
    'ALP':'ALP', 
    'IND':'IND', 
    'GRN':'GRN',
    'LP':'LIB',
    'NP':'NAT',
    'CLP':'CLP',
    'LNP':'LNP',
    'XEN':'IND',
    'KAP':'KAP',
}

modeled_parties = ['UAP', 'ONP', 'ALP', 'IND', 'GRN', 'LIB', 'NAT', 'CLP', 'LNP', 'IND', 'KAP', 'TEAL', 'OTH']

for idx,teal in teal_candidates.iterrows():
    givenNm = teal['GivenNm']
    surname = teal['Surname']
    division = teal['DivisionNm']
    primaryvotes.loc[(primaryvotes[['GivenNm','Surname','DivisionNm']]==[givenNm,surname,division]).all(axis=1),'PartyAb']="TEAL"
    if (primaryvotes[['GivenNm','Surname','DivisionNm']]==[givenNm,surname,division]).all(axis=1).sum()!=1:
        print(f"Error in {teal}")
        print(primaryvotes[(primaryvotes[['GivenNm','Surname','DivisionNm']]==[givenNm,surname,division]).all(axis=1)])
```

```python
def partymap(party):
    if party in modeled_parties:
        return party
    elif party in party_mappers.keys():
        return party_mappers[party]

    elif pd.isna(party):
        return "Informal"
    else:
        return "OTH"

primaryvotes_cleaned = primaryvotes.drop(primaryvotes[primaryvotes['Surname']=='Informal'].index)[["DivisionNm","PartyAb","TotalVotes"]].copy()

primaryvotes_cleaned['PartyAb'] = primaryvotes_cleaned['PartyAb'].apply(partymap)

```

```python
primary_votes_handled = primaryvotes_cleaned.groupby(['DivisionNm','PartyAb']).aggregate("sum").unstack()
```

```python
primary_votes_handled = primary_votes_handled.div(primary_votes_handled.sum(axis=1),axis=0)
```

```python
primary_votes_handled.columns = primary_votes_handled.columns.droplevel(0)
```

```python
overall_df = pd.concat([primary_votes_handled,demographic_db],axis=1)
```

```python
corrmat = overall_df.corr()
```
