```python
import os,sys
import pandas as pd
```

```python
demographics_folder = "2021_GCP_CED_for_AUS_short-header/2021 Census GCP Commonwealth Electroral Division for AUS"
demographic_files = os.listdir(demographics_folder)
demographic_db = pd.read_csv(f"{demographics_folder}/{demographic_files[0]}",index_col="CED_CODE_2021")
cedcodemapper = pd.read_csv("CEDcodemapper.csv",index_col="CED_CODE_2021")
demographic_db.head()
```

```python
for file in demographic_files[1:]:
    demographic_db = demographic_db.combine_first(pd.read_csv(f"{demographics_folder}/{file}",index_col="CED_CODE_2021"))

demographic_db = pd.merge(cedcodemapper,demographic_db, left_index=True, right_index=True, how='inner').set_index("DivisionNm").sort_index()
demographic_db.head()
```

```python
demographic_db_polled_variables = pd.DataFrame(
    columns=['100to150k', '150kplus', '18to34', '35to49', '50to64', '50to99k', '65plus', 'christian', 'englishonly', 'female', 'male', 'nonenglish', 'noreligion', 'notertiary', 'tafe', 'university', 'upto50k'],
    index=demographic_db.index
)
```

```python
demographic_db_polled_variables['male'] = demographic_db['Tot_P_M']/demographic_db['Tot_P_P']
demographic_db_polled_variables['female'] = demographic_db['Tot_P_F']/demographic_db['Tot_P_P']
```
